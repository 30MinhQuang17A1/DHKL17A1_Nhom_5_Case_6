Cell 1 – Import & đọc dữ liệu

```python
import pandas as pd

candidate = pd.read_csv("../data_clean/cleaned_candidate_profile.csv")
interview = pd.read_csv("../data_clean/cleaned_interview_score.csv")
salary = pd.read_csv("../data_clean/cleaned_salary_offer.csv")
```

Cell 2 – Kiểm tra dữ liệu

```python
candidate.head()
interview.head()
salary.head()

```

Cell 3 – Merge candidate + interview

```python
df = candidate.merge(
    interview,
    on="candidate_id",
    how="left"
)

df.head()
```

Cell 4 – Merge thêm salary

```python
df = df.merge(
    salary,
    on="candidate_id",
    how="left"
)

df.head()
```

Cell 5 – Phát hiện mismatch

```python
mismatch = df[df.isna().any(axis=1)]
mismatch
```

Cell 6 – Xuất file hoàn chỉnh

```python
df.to_csv("../data_clean/full_hr_data.csv", index=False)
```
